```python
# Define function to calculate technical indicators
def calculate_technical_indicators(df, close_col='close', window=14):
    """Calculate RSI and MACD for a given dataframe."""
    # Copy dataframe to avoid modifying original
    df = df.copy().sort_values('date')
    
    # Calculate daily returns
    df['returns'] = df[close_col].pct_change()
    
    # Calculate RSI
    df['gain'] = np.where(df['returns'] > 0, df['returns'], 0)
    df['loss'] = np.where(df['returns'] < 0, -df['returns'], 0)
    
    # Calculate average gain and average loss
    avg_gain = df['gain'].rolling(window=window).mean()
    avg_loss = df['loss'].rolling(window=window).mean()
    
    # Calculate RS and RSI
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    
    # Calculate MACD
    df['EMA12'] = df[close_col].ewm(span=12, adjust=False).mean()
    df['EMA26'] = df[close_col].ewm(span=26, adjust=False).mean()
    df['MACD'] = df['EMA12'] - df['EMA26']
    df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['MACD_Histogram'] = df['MACD'] - df['Signal_Line']
    
    return df

# Apply technical indicators to selected stock
if single_stock_df is not None:
    if 'close' in single_stock_df.columns:
        single_stock_df = calculate_technical_indicators(single_stock_df)
        print("\nTechnical indicators calculated successfully")
        print(single_stock_df[['date', 'close', 'RSI', 'MACD', 'Signal_Line']].head(10))
    else:
        # Check for alternative close price columns
        print("\nCould not find 'close' column for technical indicators")
        print(f"Available columns: {single_stock_df.columns.tolist()}")
        
        # Look for potential close price columns
        potential_close_cols = [col for col in single_stock_df.columns if 'close' in col.lower() or 'adj' in col.lower() or 'price' in col.lower()]
        
        if potential_close_cols:
            print(f"\nPotential close price columns found: {potential_close_cols}")
            close_col = potential_close_cols[0]
            print(f"Using '{close_col}' as close price column")
            single_stock_df = calculate_technical_indicators(single_stock_df, close_col=close_col)
            print("Technical indicators calculated successfully")
            print(single_stock_df[['date', close_col, 'RSI', 'MACD', 'Signal_Line']].head(10))
else:
    print("\nNo stock data selected for technical indicators")
```

```python
# Import necessary libraries for sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon if needed
try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except LookupError:
    nltk.download('vader_lexicon')

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
    """Analyze sentiment using VADER."""
    if pd.isna(text):
        return {'compound': 0, 'neg': 0, 'neu': 0, 'pos': 0}
    return sia.polarity_scores(text)

# Get the headline columns (Top1, Top2, etc.)
top_cols = [col for col in news_df.columns if col.startswith('Top')]
print(f"\nFound {len(top_cols)} headline columns: {top_cols[:5]}...")

if top_cols:
    print("\nPerforming sentiment analysis on all headlines...")
    
    # Create an empty DataFrame to store daily sentiment
    daily_sentiment = pd.DataFrame()
    daily_sentiment['date'] = news_df['date'].unique()
    daily_sentiment['compound'] = 0.0
    daily_sentiment['negative'] = 0.0
    daily_sentiment['neutral'] = 0.0
    daily_sentiment['positive'] = 0.0
    
    # Keep track of headline counts for averaging
    daily_sentiment['headline_count'] = 0
    
    # Process each headline column
    for col in top_cols:
        print(f"Processing {col}...")
        
        # Apply sentiment analysis to this headline column
        sentiments = news_df[col].apply(analyze_sentiment)
        
        # Extract sentiment scores into temporary columns
        news_df[f'{col}_compound'] = sentiments.apply(lambda x: x['compound'])
        news_df[f'{col}_negative'] = sentiments.apply(lambda x: x['neg'])
        news_df[f'{col}_neutral'] = sentiments.apply(lambda x: x['neu'])
        news_df[f'{col}_positive'] = sentiments.apply(lambda x: x['pos'])
        
        # Print a sample of the sentiment scores
        if col == top_cols[0]:  # Just show for the first column
            print("\nSample sentiment scores for first headline column:")
            print(news_df[[col, f'{col}_compound', f'{col}_negative', f'{col}_neutral', f'{col}_positive']].head())
    
    # Now aggregate the sentiment across all headline columns for each date
    print("\nAggregating sentiment across all headlines by date...")
    
    for date in daily_sentiment['date']:
        date_rows = news_df[news_df['date'] == date]
        
        for col in top_cols:
            # Only include non-NaN headlines
            valid_headlines = date_rows[~date_rows[col].isna()]
            headline_count = len(valid_headlines)
            
            if headline_count > 0:
                # Add sentiment scores for this column to the daily totals
                index = daily_sentiment[daily_sentiment['date'] == date].index[0]
                daily_sentiment.loc[index, 'compound'] += valid_headlines[f'{col}_compound'].sum()
                daily_sentiment.loc[index, 'negative'] += valid_headlines[f'{col}_negative'].sum()
                daily_sentiment.loc[index, 'neutral'] += valid_headlines[f'{col}_neutral'].sum()
                daily_sentiment.loc[index, 'positive'] += valid_headlines[f'{col}_positive'].sum()
                daily_sentiment.loc[index, 'headline_count'] += headline_count
    
    # Calculate average sentiment per day
    daily_sentiment['compound'] = daily_sentiment.apply(
        lambda row: row['compound'] / row['headline_count'] if row['headline_count'] > 0 else 0, axis=1
    )
    daily_sentiment['negative'] = daily_sentiment.apply(
        lambda row: row['negative'] / row['headline_count'] if row['headline_count'] > 0 else 0, axis=1
    )
    daily_sentiment['neutral'] = daily_sentiment.apply(
        lambda row: row['neutral'] / row['headline_count'] if row['headline_count'] > 0 else 0, axis=1
    )
    daily_sentiment['positive'] = daily_sentiment.apply(
        lambda row: row['positive'] / row['headline_count'] if row['headline_count'] > 0 else 0, axis=1
    )
    
    # Drop the headline_count column as it's no longer needed
    daily_sentiment = daily_sentiment.drop(columns=['headline_count'])
    
    print("\nDaily sentiment aggregation completed")
    print(daily_sentiment.head())
    
    # Save daily sentiment for reference
    daily_sentiment.to_csv('daily_news_sentiment.csv', index=False)
    print("\nDaily sentiment saved to 'daily_news_sentiment.csv'")
else:
    print("\nNo headline columns identified in the news dataset")
```

```python
# Import necessary libraries for modeling
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import pickle

# Check if merged data is available
if 'merged_df' in locals() and not merged_df.empty:
    # Prepare for modeling
    merged_df = merged_df.dropna()
    
    # Identify close price column
    if 'close' in merged_df.columns:
        close_col = 'close'
    else:
        potential_close_cols = [col for col in merged_df.columns if 'close' in col.lower() or 'adj' in col.lower() or 'price' in col.lower()]
        close_col = potential_close_cols[0] if potential_close_cols else None
    
    if close_col is not None:
        print(f"\nUsing '{close_col}' as the target price column for prediction")
        
        # Define features
        # Technical indicators
        tech_features = ['RSI', 'MACD', 'Signal_Line', 'EMA12', 'EMA26', 'MACD_Histogram']
        # Sentiment features (either actual or placeholder)
        sentiment_features = ['compound', 'negative', 'positive']
        
        # Check which features are available
        tech_available = [f for f in tech_features if f in merged_df.columns]
        sentiment_available = [f for f in sentiment_features if f in merged_df.columns]
        
        available_features = tech_available + sentiment_available
        
        print(f"\nAvailable technical indicators: {tech_available}")
        print(f"Available sentiment features: {sentiment_available}")
        print(f"Total features for modeling: {len(available_features)}")
        
        if len(available_features) >= 2:  # Need at least two features
            # Create feature set
            X = merged_df[available_features]
            
            # Create target - next day's closing price
            merged_df['next_day_close'] = merged_df[close_col].shift(-1)
            y_reg = merged_df['next_day_close'].dropna()
            
            # Align features with target
            X_reg = X.iloc[:len(y_reg)]
            
            print(f"\nPrepared {len(X_reg)} data points for regression modeling")
            
            # Split data into training and testing sets
            X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
                X_reg, y_reg, test_size=0.2, random_state=42, shuffle=False
            )
            
            print(f"Training set: {len(X_train_reg)} samples")
            print(f"Testing set: {len(X_test_reg)} samples")
            
            # Standardize features
            scaler = StandardScaler()
            X_train_reg_scaled = scaler.fit_transform(X_train_reg)
            X_test_reg_scaled = scaler.transform(X_test_reg)
            
            # Train Linear Regression model
            lr_model = LinearRegression()
            lr_model.fit(X_train_reg_scaled, y_train_reg)
            
            # Make predictions
            y_pred_reg = lr_model.predict(X_test_reg_scaled)
            
            # Evaluate model
            mse = mean_squared_error(y_test_reg, y_pred_reg)
            mae = mean_absolute_error(y_test_reg, y_pred_reg)
            rmse = np.sqrt(mse)
            
            print("\nLinear Regression Results:")
            print(f"Mean Squared Error: {mse:.4f}")
            print(f"Mean Absolute Error: {mae:.4f}")
            print(f"Root Mean Squared Error: {rmse:.4f}")
            
            # Feature importance
            lr_importance = pd.DataFrame({
                'Feature': available_features,
                'Importance': np.abs(lr_model.coef_)
            })
            lr_importance = lr_importance.sort_values('Importance', ascending=False)
            print("\nLinear Regression Feature Importance:")
            print(lr_importance)
            
            # Save model and feature importance
            with open('linear_regression_model.pkl', 'wb') as f:
                pickle.dump(lr_model, f)
            lr_importance.to_csv('linear_regression_importance.csv', index=False)
            
            print("\nLinear Regression model and feature importance saved")
            
            # Visualize actual vs predicted prices
            plt.figure(figsize=(12, 6))
            plt.plot(y_test_reg.index, y_test_reg.values, label='Actual')
            plt.plot(y_test_reg.index, y_pred_reg, label='Predicted', color='red')
            plt.title('Linear Regression: Actual vs Predicted Prices')
            plt.legend()
            plt.tight_layout()
            plt.savefig('linear_regression_prediction.png')
            print("\nLinear Regression prediction plot saved to 'linear_regression_prediction.png'")
        else:
            print(f"\nNot enough features available for modeling. Found: {available_features}")
    else:
        print("\nNo suitable close price column found for regression modeling")
else:
    print("\nNo processed data available for modeling")
```

```python
# Import for Random Forest model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Check if we have processed data and close price column
if 'merged_df' in locals() and not merged_df.empty and 'close_col' in locals() and close_col is not None:
    # For classification - predict trend (up/down)
    merged_df['price_change'] = merged_df['next_day_close'] - merged_df[close_col]
    merged_df['trend'] = np.where(merged_df['price_change'] > 0, 1, 0)  # 1 for up, 0 for down
    y_clf = merged_df['trend'].dropna()
    
    print(f"\nPreparing classification task: Predicting market trend (up/down)")
    print(f"Class distribution: Up={sum(y_clf == 1)} ({sum(y_clf == 1)/len(y_clf):.1%}), Down={sum(y_clf == 0)} ({sum(y_clf == 0)/len(y_clf):.1%})")
    
    # Align features with target
    X_clf = X.iloc[:len(y_clf)]
    
    # Split data
    X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
        X_clf, y_clf, test_size=0.2, random_state=42, shuffle=False
    )
    
    print(f"Training set: {len(X_train_clf)} samples")
    print(f"Testing set: {len(X_test_clf)} samples")
    
    # Standardize features
    X_train_clf_scaled = scaler.fit_transform(X_train_clf)
    X_test_clf_scaled = scaler.transform(X_test_clf)
    
    # Train Random Forest Classifier
    rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
    rf_model.fit(X_train_clf_scaled, y_train_clf)
    
    # Make predictions
    y_pred_clf = rf_model.predict(X_test_clf_scaled)
    
    # Evaluate model
    print("\nRandom Forest Classification Results:")
    print(classification_report(y_test_clf, y_pred_clf))
    
    # Confusion matrix
    cm = confusion_matrix(y_test_clf, y_pred_clf)
    print("\nConfusion Matrix:")
    print(cm)
    
    # Feature importance
    rf_importance = pd.DataFrame({
        'Feature': available_features,
        'Importance': rf_model.feature_importances_
    })
    rf_importance = rf_importance.sort_values('Importance', ascending=False)
    print("\nRandom Forest Feature Importance:")
    print(rf_importance)
    
    # Save model and feature importance
    with open('random_forest_model.pkl', 'wb') as f:
        pickle.dump(rf_model, f)
    rf_importance.to_csv('random_forest_importance.csv', index=False)
    
    print("\nRandom Forest model and feature importance saved")
    
    # Visualize feature importance
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=rf_importance)
    plt.title('Random Forest Feature Importance')
    plt.tight_layout()
    plt.savefig('random_forest_importance.png')
    print("\nRandom Forest feature importance plot saved to 'random_forest_importance.png'")
    
    # Visualize confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix for Random Forest')
    plt.tight_layout()
    plt.savefig('random_forest_confusion_matrix.png')
    print("\nConfusion matrix visualization saved to 'random_forest_confusion_matrix.png'")
else:
    print("\nCannot perform classification modeling - required data not available")
```

```python
# Import libraries for ARIMA modeling
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
warnings.filterwarnings('ignore')

# Make sure we have processed stock data available
if 'merged_df' in locals() and not merged_df.empty and close_col is not None:
    print("\n===== ARIMA Modeling =====")
    
    # Select only the relevant time series data
    ts_data = merged_df[['date', close_col]].copy()
    ts_data = ts_data.sort_values('date')
    ts_data = ts_data.set_index('date')
    
    # Check for stationarity using Augmented Dickey-Fuller test
    print("\nPerforming stationarity test on price data:")
    adf_result = adfuller(ts_data[close_col].dropna())
    print(f'ADF Statistic: {adf_result[0]:.4f}')
    print(f'p-value: {adf_result[1]:.4f}')
    print(f'Critical Values: {adf_result[4]}')
    
    if adf_result[1] > 0.05:
        print("Time series is non-stationary. Differencing will be needed.")
        # Calculate first difference
        ts_data['price_diff'] = ts_data[close_col].diff().dropna()
        
        # Check stationarity of differenced series
        adf_diff_result = adfuller(ts_data['price_diff'].dropna())
        print("\nAfter differencing:")
        print(f'ADF Statistic: {adf_diff_result[0]:.4f}')
        print(f'p-value: {adf_diff_result[1]:.4f}')
        
        # Use differenced series if it's stationary
        if adf_diff_result[1] <= 0.05:
            print("Differenced series is stationary.")
            arima_series = ts_data['price_diff'].dropna()
            d_value = 1
        else:
            print("Differenced series is still non-stationary. Using second differencing.")
            ts_data['price_diff2'] = ts_data['price_diff'].diff().dropna()
            arima_series = ts_data['price_diff2'].dropna()
            d_value = 2
    else:
        print("Time series is already stationary.")
        arima_series = ts_data[close_col].dropna()
        d_value = 0
    
    # Plot ACF and PACF to help identify p and q values
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plot_acf(arima_series, ax=plt.gca(), lags=20)
    plt.title('ACF')
    
    plt.subplot(122)
    plot_pacf(arima_series, ax=plt.gca(), lags=20)
    plt.title('PACF')
    plt.tight_layout()
    plt.savefig('acf_pacf_plots.png')
    print("\nACF and PACF plots saved to 'acf_pacf_plots.png'")
    
    # Split data into train and test sets
    train_size = int(len(arima_series) * 0.8)
    train_data = arima_series[:train_size]
    test_data = arima_series[train_size:]
    
    print(f"\nTraining ARIMA on {len(train_data)} data points, testing on {len(test_data)} data points")
    
    # Try a few different ARIMA models
    print("\nFitting ARIMA models...")
    p_values = [1, 2]
    q_values = [0, 1]
    best_aic = float('inf')
    best_model = None
    best_params = None
    
    for p in p_values:
        for q in q_values:
            try:
                model = ARIMA(train_data, order=(p, d_value, q))
                model_fit = model.fit()
                aic = model_fit.aic
                
                print(f"ARIMA({p},{d_value},{q}) - AIC: {aic:.4f}")
                
                if aic < best_aic:
                    best_aic = aic
                    best_model = model_fit
                    best_params = (p, d_value, q)
            except Exception as e:
                print(f"Error fitting ARIMA({p},{d_value},{q}): {e}")
    
    if best_model is not None:
        print(f"\nBest ARIMA model: ARIMA{best_params} with AIC: {best_aic:.4f}")
        
        # Make predictions with the best model
        # Forecast on test set
        forecast_steps = len(test_data)
        forecast = best_model.forecast(steps=forecast_steps)
        
        # If we differenced, we need to transform forecasts back to original scale
        if d_value >= 1:
            # For simplicity, just compare the differenced values
            print("\nNote: Predictions are on differenced scale")
            
            # Calculate error metrics on the differenced scale
            mse_arima = mean_squared_error(test_data, forecast)
            rmse_arima = np.sqrt(mse_arima)
            mae_arima = mean_absolute_error(test_data, forecast)
            
            print("\nARIMA Forecast Results (on differenced scale):")
            print(f"Mean Squared Error: {mse_arima:.4f}")
            print(f"Root Mean Squared Error: {rmse_arima:.4f}")
            print(f"Mean Absolute Error: {mae_arima:.4f}")
        else:
            # If no differencing, metrics are on original scale
            mse_arima = mean_squared_error(test_data, forecast)
            rmse_arima = np.sqrt(mse_arima)
            mae_arima = mean_absolute_error(test_data, forecast)
            
            print("\nARIMA Forecast Results:")
            print(f"Mean Squared Error: {mse_arima:.4f}")
            print(f"Root Mean Squared Error: {rmse_arima:.4f}")
            print(f"Mean Absolute Error: {mae_arima:.4f}")
        
        # Plot actual vs forecast
        plt.figure(figsize=(12, 6))
        plt.plot(test_data.index, test_data, label='Actual')
        plt.plot(test_data.index, forecast, label='Forecast', color='red')
        plt.title(f'ARIMA{best_params} Forecast vs Actual')
        plt.legend()
        plt.tight_layout()
        plt.savefig('arima_forecast.png')
        print("\nARIMA forecast plot saved to 'arima_forecast.png'")
        
        # Try to save the ARIMA model - using joblib instead of pickle for better compatibility
        try:
            import joblib
            joblib.dump(best_model, 'arima_model.joblib')
            print("\nBest ARIMA model saved to 'arima_model.joblib'")
        except Exception as e:
            print(f"\nError saving ARIMA model: {e}")
            print("This is not critical - ARIMA models can be difficult to serialize")
        
        # Compare ARIMA with Linear Regression
        if 'rmse' in locals():  # Check if Linear Regression results exist
            print("\nModel Comparison:")
            print(f"ARIMA RMSE: {rmse_arima:.4f}")
            print(f"Linear Regression RMSE: {rmse:.4f}")
            
            if rmse_arima < rmse:
                print("ARIMA performs better than Linear Regression for this data")
            else:
                print("Linear Regression performs better than ARIMA for this data")
    else:
        print("\nNo valid ARIMA model could be fitted to the data")
else:
    print("\nCannot perform ARIMA modeling - stock data not available")
```
