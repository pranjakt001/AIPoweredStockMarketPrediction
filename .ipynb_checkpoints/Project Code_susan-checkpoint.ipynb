```python
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Sort by ticker and date
df.sort_values(by=['Name', 'Date'], inplace=True)

# Visualize a specific stock's trend (e.g., CBA)
sample_name = 'CBA'
cba_df = df[df['Name'] == sample_name]

plt.figure(figsize=(10, 5))
plt.plot(cba_df['Date'], cba_df['Close'])
plt.title(f"{sample_name} Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter for stocks with full data (optional: e.g. > 1000 rows)
stock_counts = df['Name'].value_counts()
valid_stocks = stock_counts[stock_counts > 1000].index.tolist()
df_filtered = df[df['Name'].isin(valid_stocks)]

# Group by stock and compute features
features = []

for name, group in df_filtered.groupby('Name'):
    group = group.sort_values('Date')
    
    # Volatility: standard deviation of closing price
    volatility = group['Close'].std()
    
    # Average volume
    avg_volume = group['Volume'].mean()
    
    # Average closing price
    avg_close = group['Close'].mean()
    
    # Recent price momentum (last 30 days)
    recent = group[group['Date'] >= group['Date'].max() - pd.Timedelta(days=30)]
    if len(recent) > 5:  # To avoid unstable ratios
        momentum = (recent['Close'].iloc[-1] - recent['Close'].iloc[0]) / recent['Close'].iloc[0]
    else:
        momentum = 0
    
    features.append({
        'Name': name,
        'Volatility': volatility,
        'AvgVolume': avg_volume,
        'AvgClose': avg_close,
        'Momentum30d': momentum
    })

# Convert to DataFrame
feature_df = pd.DataFrame(features)

# Normalize features for fair comparison (optional)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalized = scaler.fit_transform(feature_df[['Volatility', 'AvgVolume', 'AvgClose', 'Momentum30d']])
feature_df[['Volatility_norm', 'Volume_norm', 'Close_norm', 'Momentum_norm']] = normalized

# Apply custom weights (Explainability-focused)
weights = {
    'Volatility_norm': 0.4,
    'Volume_norm': 0.2,
    'Close_norm': 0.2,
    'Momentum_norm': 0.2
}

# Calculate weighted score
feature_df['WeightedScore'] = (
    feature_df['Volatility_norm'] * weights['Volatility_norm'] +
    feature_df['Volume_norm'] * weights['Volume_norm'] +
    feature_df['Close_norm'] * weights['Close_norm'] +
    feature_df['Momentum_norm'] * weights['Momentum_norm']
)


# Sort by score and get top 5
top_weighted_stocks = feature_df.sort_values(by='WeightedScore', ascending=False).head(5)

# Show results
print(" Top 5 stocks selected based on weighted practical criteria:")
print(top_weighted_stocks[['Name', 'Volatility', 'AvgVolume', 'AvgClose', 'Momentum30d', 'WeightedScore']])
```

## Add Technical Indicators (SMA, EMA, RSI, MACD)

* Capture overbought/oversold signals and momentum shift

* Great features for interpretable models 

**Technical Indicator Descriptions**

| Column Name   | Description |
|---------------|-------------|
| `SMA_20`      | 20-day Simple Moving Average  the average of the closing prices over the past 20 trading days |
| `EMA_20`      | 20-day Exponential Moving Average  a weighted average that gives more importance to recent prices |
| `RSI_14`      | 14-day Relative Strength Index  measures momentum by comparing recent gains and losses |
| `MACD`        | Moving Average Convergence Divergence  calculated as the difference between the 12-day and 26-day EMAs |
| `MACD_signal` | Signal line for MACD  the 9-day EMA of the MACD line, used to identify buy/sell signals |

```python
import numpy as np
```

```python
# Calculate price difference
delta = asx_index['Close'].diff()

# Compute gains and losses (ensure 1D with .flatten())
gain = np.where(delta > 0, delta, 0).flatten()
loss = np.where(delta < 0, -delta, 0).flatten()

# Convert back to Series with correct index
gain_series = pd.Series(gain, index=asx_index.index)
loss_series = pd.Series(loss, index=asx_index.index)

# Compute rolling averages (14-day)
avg_gain = gain_series.rolling(window=14).mean()
avg_loss = loss_series.rolling(window=14).mean()

# Compute RSI
rs = avg_gain / avg_loss
asx_index['RSI_14'] = 100 - (100 / (1 + rs))
```

```python
# Add SMA, EMA (if not already added)
asx_index['SMA_20'] = asx_index['Close'].rolling(window=20).mean()
asx_index['EMA_20'] = asx_index['Close'].ewm(span=20, adjust=False).mean()

# MACD and Signal Line (if not already added)
ema_12 = asx_index['Close'].ewm(span=12, adjust=False).mean()
ema_26 = asx_index['Close'].ewm(span=26, adjust=False).mean()
asx_index['MACD'] = ema_12 - ema_26
asx_index['MACD_signal'] = asx_index['MACD'].ewm(span=9, adjust=False).mean()

# Preview last 5 rows of technical indicators
print(" Technical indicators added (ASX200 index):")
print(asx_index[['Date', 'Close', 'SMA_20', 'EMA_20', 'RSI_14', 'MACD', 'MACD_signal']].tail())
```

Additionally, MACD is significantly below the signal line, confirming strong bearish momentum.

# Data Preprocessing & Input Construction

## Objective
Use the past **N day** of time-series as input **X**, and predict either the **closing price D days ahead** or the **price direction (up/down)** as output **Y**.

## Example Setting
- `N = 30`: Use the past 30 days of data as input
- `D = 5`: Predict the stock/index behavior 5 days into the future

## Required Features
- **Close** price
- **Optional technical indicators**:
    - RSI (Relative Strength Index)
    - MACD (Moving Average Convergence Divergence)
    - SMA (Simple Moving Average)
    - EMA (Exponential Moving Average) 

```python
print(" Missing values per column:")
print(asx_index.isna().sum())
```

```python
# Drop any rows that contain NaN values (usually from early days of technical indicator calculations)
asx_index_cleaned = asx_index.dropna()

# Check the result
print(f" Cleaned dataset shape: {asx_index_cleaned.shape}")
print(asx_index_cleaned.head())
```

```python
# Safely convert Date to datetime using .loc
asx_index_cleaned.loc[:, 'Date'] = pd.to_datetime(asx_index_cleaned['Date'])

# Set 'Date' as index
asx_index_cleaned = asx_index_cleaned.set_index('Date')

# Sort by index (date)
asx_index_cleaned = asx_index_cleaned.sort_index()
```

```python
# Input Features
features = ['Close', 'RSI_14', 'MACD', 'MACD_signal', 'SMA_20', 'EMA_20']

# Sliding Window Input/Output
# Parameters
N = 30  # number of past days (window)
D = 5   # days ahead to predict

# Convert to numpy array for performance
data = asx_index_cleaned[features].values
close_prices = asx_index_cleaned['Close'].values  # for target

# Store input-output pairs
X = []
y_regression = []  # For regression: future price
y_classification = []  # For classification: up/down

for i in range(len(data) - N - D):
    X.append(data[i:i+N])  # N-day input
    future_close = close_prices[i + N + D - 1]
    current_close = close_prices[i + N - 1]
    
    y_regression.append(future_close)  # Regression target
    y_classification.append(1 if future_close > current_close else 0)  # Classification target

# Convert to numpy arrays
X = np.array(X)
y_regression = np.array(y_regression)
y_classification = np.array(y_classification)

print(" Input shape (X):", X.shape)  # (samples, N, features)
print(" Regression target shape:", y_regression.shape)
print(" Classification target shape:", y_classification.shape)
```

# Model Design & Experimentation

## Baseline Model
### 1. Classification- Logistic Regression

Used as a baseline for performance comparison between traditional and deep learning models.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Flatten input (30 days  6 features  180)
X_flattened = X.reshape(X.shape[0], -1)

# 2. Time-aware train/test split (no shuffling)
split_idx = int(len(X_flattened) * 0.8)
X_train, X_test = X_flattened[:split_idx], X_flattened[split_idx:]
y_train, y_test = y_classification[:split_idx], y_classification[split_idx:]

# 3. Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Train baseline model (Logistic Regression)
baseline_model = LogisticRegression(max_iter=1000)
baseline_model.fit(X_train_scaled, y_train)

# 5. Predict and evaluate
y_pred = baseline_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f" Accuracy: {accuracy:.4f}")
print(f" F1 Score: {f1:.4f}")
print("\n Detailed Report:\n", classification_report(y_test, y_pred))
```

* The model correctly predicted about **49% of the cases**.
* F1 Score: 0.3881 indicates **low predictive performance**, especially considering class imbalance.
* For class 0 (down): The model **predicted many correctly**, but with **low precision**.
* For class 1 (up): The model was **precise when it predicted "up"**, but **missed most cases** (low recall).

* A simple linear model like Logistic Regression struggles to capture complex temporal patterns in time series data.
    * Logistic regression is sensitive to distance-based calculations, so if the features have different scales, it can negatively affect the model's performance. That's why standardises them to have (mean = 0, standard deviation = 1). 
* The model may be underfitting, as it lacks capacity to learn from limited technical indicators.
* Market behavior is inherently complex, and using only a few features (e.g., RSI, MACD, etc.) might not be sufficient.

## Deep Learning Model

**GRU**

- Learns patterns from continuous data flow

- Lightweight and faster alternative to traditional RNN

- Predicts up/down movements based on past 30-day patterns

**LSTM**

- Capable of capturing long-term dependencies

- Suitable for complex time series

 We proceed with the faster and lighter GRU model.

### 1. **GRU + SHAP**

* Framework: TensorFlow + Keras  Simple and clean syntax, with good SHAP compatibility

* Task: Binary classification (predicting price up or down)

* Model: GRU  To capture temporal dependencies in time-series data, followed by SHAP for explainability

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

#### GRU Model Training for Binary Classification

```python
# Reshape from (samples, 180)  (samples, 30, 6)
X_train_gru = X_train_scaled.reshape(-1, 30, 6)
X_test_gru = X_test_scaled.reshape(-1, 30, 6)
```

```python
# 1. Define the GRU model
model = Sequential([
    GRU(64, input_shape=(30, 6), return_sequences=False),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')  # Binary classification (up/down)
])

# 2. Compile the model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 3. Train the model
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train_gru, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# 4. Evaluate on test set
test_loss, test_acc = model.evaluate(X_test_gru, y_test)
print(f"\n Test Accuracy: {test_acc:.4f}")
```

### Why did GRU underperform?
- **Underfitting**: The model may not have learned enough due to a shallow structure or too few epochs.
- **Hyperparameters**: GRU unit size, learning rate, or optimizer may need tuning. 
- **Feature scaling or selection**: The input features might be insufficient or contain too much noise. 
- **Evaluation method**: Accuracy alone may be misleading  F1 score and confusion matrix provide better insights. 

### Suggested Next Steps

- **Increase training epochs** (e.g., currently it trains for only 56 epochs).
- **Improve model architecture**:
  - Increase the number of GRU units (e.g., 32  64).
  - Add **Dropout layers** for regularization.
  - Implement **EarlyStopping** and **ModelCheckpoint** to avoid overfitting and save the best model.

```python
# 1. Build the model
model = Sequential([
    GRU(64, input_shape=(30, 6), return_sequences=False),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 2. Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_gru_model.h5', monitor='val_loss',
                             save_best_only=True, verbose=1)

# 3. Train the model
history = model.fit(
    X_train_gru, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, checkpoint],
    verbose=1
)

# 4. Evaluate
test_loss, test_acc = model.evaluate(X_test_gru, y_test, verbose=0)
print(f"\n Improved GRU Test Accuracy: {test_acc:.4f}")
```
